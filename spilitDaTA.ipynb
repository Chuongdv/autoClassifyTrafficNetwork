```python
import pandas as pd
import random
```

```python
dataframe = pd.read_csv('./data/trace2/trace2.csv', encoding='utf-8', header=0, sep=',', index_col=0)
```

```python
dataframe.drop(['Source_IP', 'Source_Port', 'Destination_IP', 'Destination_Port', 'Protocol'], axis = 1, inplace = True)
```

```python
dataframe.info()
```

```python
testSet = []
def spilit(x):
    testSet.append(x.sample(frac=0.2,random_state=200))
dataframe.groupby(['ProtocolName']).apply(spilit)
resultTest = pd.concat(testSet)
```

```python
resultTest.info()
```

```python
resultTest.groupby(['ProtocolName']).count()
```

```python
dataframe.groupby(['ProtocolName']).count()
```

```python
dataframe = dataframe.drop(resultTest.index)
```

```python
resultTest.reset_index(drop=True, inplace = True)
dataframe.reset_index(drop=True, inplace = True)
```
