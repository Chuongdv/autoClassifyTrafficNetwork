```python
# Render our plots inline
%matplotlib inline
 
import pandas as pd
import matplotlib.pyplot as plt
import random

dataframe = pd.read_csv('./data/trace1/trace1.csv', encoding="utf-8", header=0, sep=',', index_col=0)
dataframe.head(5)
```

```python
dataframe.info()
```

```python
dataframe.groupby(['label']).count()
```

```python
dns = dataframe.query('label == "dns"')
ftp = dataframe.query('label == "ftp"')
http= dataframe.query('label == "http"')
https = dataframe.query('label == "https"')
imap = dataframe.query('label == "imap"')
imaps = dataframe.query('label == "imaps"')
pop3 = dataframe.query('label == "pop3"')
pop3s = dataframe.query('label == "pop3s"')
smtp = dataframe.query('label == "smtp"')
ssh = dataframe.query('label == "ssh"')
```

```python
dnsTest = dns.sample(frac=0.2,random_state=200) 
dnsTraining = dns.drop(dnsTest.index)
dnsOptimal = dnsTraining.sample(frac=0.2,random_state=200) 

ftpTest = ftp.sample(frac=0.2,random_state=200) 
ftpTraining = ftp.drop(ftpTest.index)
ftpOptimal = ftpTraining.sample(frac=0.2,random_state=200)

httpTest = http.sample(frac=0.2,random_state=200) 
httpTraining = http.drop(httpTest.index)
httpOptimal = httpTraining.sample(frac=0.2,random_state=200)

httpsTest = https.sample(frac=0.2,random_state=200) 
httpsTraining = https.drop(httpsTest.index)
httpsOptimal = httpsTraining.sample(frac=0.2,random_state=200)

imapTest = imap.sample(frac=0.2,random_state=200) 
imapTraining = imap.drop(imapTest.index)
imapOptimal = imapTraining.sample(frac=0.2,random_state=200)

imapsTest = imaps.sample(frac=0.2,random_state=200) 
imapsTraining = imaps.drop(imapsTest.index)
imapsOptimal = imapsTraining.sample(frac=0.2,random_state=200)

pop3Test = pop3.sample(frac=0.2,random_state=200) 
pop3Training = pop3.drop(pop3Test.index)
pop3Optimal = pop3Training.sample(frac=0.2,random_state=200)

pop3sTest = pop3s.sample(frac=0.2,random_state=200) 
pop3sTraining = pop3s.drop(pop3sTest.index)
pop3sOptimal = pop3sTraining.sample(frac=0.2,random_state=200)

smtpTest = smtp.sample(frac=0.2,random_state=200) 
smtpTraining = smtp.drop(smtpTest.index)
smtpOptimal = smtpTraining.sample(frac=0.2,random_state=200)

sshTest = ssh.sample(frac=0.2,random_state=200) 
sshTraining = ssh.drop(sshTest.index)
sshOptimal = sshTraining.sample(frac=0.2,random_state=200)





 
```

```python
setTest = dnsTest.append(ftpTest, ignore_index = True).append(httpTest, ignore_index = True).append(httpsTest, ignore_index = True)\
.append(imapTest, ignore_index = True).append(imapsTest, ignore_index = True).append(pop3Test, ignore_index = True)\
.append(pop3sTest, ignore_index = True).append(smtpTest, ignore_index = True).append(sshTest, ignore_index = True)
```

```python
setTrain = dnsTraining.append(ftpTraining, ignore_index = True).append(httpTraining, ignore_index = True).append(httpsTraining, ignore_index = True)\
.append(imapTraining, ignore_index = True).append(imapsTraining, ignore_index = True).append(pop3Training, ignore_index = True)\
.append(pop3sTraining, ignore_index = True).append(smtpTraining, ignore_index = True).append(sshTraining, ignore_index = True)
```

```python
setTrain.info()
```

```python
setOptimal = dnsOptimal.append(ftpOptimal, ignore_index = True).append(httpOptimal, ignore_index = True).append(httpsOptimal, ignore_index = True)\
.append(imapOptimal, ignore_index = True).append(imapsOptimal, ignore_index = True).append(pop3Optimal, ignore_index = True)\
.append(pop3sOptimal, ignore_index = True).append(smtpOptimal, ignore_index = True).append(sshOptimal, ignore_index = True)
```
