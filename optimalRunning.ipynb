```python
# Render our plots inline
%matplotlib inline
 
import pandas as pd
import matplotlib.pyplot as plt
import random

optimalFrame = pd.read_csv('./data/trace2/optimalSet.csv', encoding="utf-8", header=0, sep=',', index_col=0)
optimalFrame.head(5)
```

```python
datarun = optimalFrame[['Fwd_Packet_Length_Mean','Fwd_Packet_Length_Std',\
                    'Bwd_Packet_Length_Std', 'Bwd_Packet_Length_Mean', 'Fwd_IAT_Mean', 'Flow_Duration', 'Total_Length_of_Fwd_Packets']]
```

```python
datarun.index.name = "index"
datarun.to_csv('./data/trace2/autoclassOptimal.tsv',sep="\t", index = True, header = True)
```

```python
import autoclasswrapper as wrapper
# Create object to prepare dataset.
clust = wrapper.Input()

# Load datasets from tsv files.
clust.add_input_data("./data/trace2/autoclassOptimal.tsv", "real location")

# Prepare input data:
# - create a final dataframe
# - merge datasets if multiple inputs
clust.prepare_input_data()

# Create files needed by AutoClass.
clust.create_db2_file()
clust.create_hd2_file()
clust.create_model_file()
clust.create_sparams_file(max_duration = 0)
clust.create_rparams_file()
```

```python
# Clean previous status file and results if a classification has already been performed.
!rm -f autoclass-run-* *.results-bin

# Search autoclass in path.
wrapper.search_autoclass_in_path()

# Create object to run AutoClass.
run = wrapper.Run()

# Prepare run script.
run.create_run_file()

# Run AutoClass.
run.run()
```
